```python
import pandas as pd
data=pd.read_csv("/kaggle/input/retail-store-sales-dirty-for-data-cleaning/retail_store_sales.csv")
data
```

**Rename Multiple columns**

```python
'''Item                 True
Price Per Unit       True
Quantity             True
Total Spent '''
'''x= data['Item'].fillna('unknown')
x
x.isnull().any()'''

data['Item'].fillna('unknown',inplace=True)
print("Item : ",data['Item'].isnull().any())

data['Price Per Unit'].fillna(data['Price Per Unit'].mean,inplace=True)
print("Price Per Unit :",data['Price Per Unit'].isnull().any())

data['Quantity'].fillna(data['Quantity'].median,inplace=True)
print("Quantity :",data['Quantity'].isnull().any())

data['Total Spent'].fillna(data['Total Spent'].median,inplace=True)
print("Total Spent :",data['Total Spent'].isnull().any())

data['Discount Applied'].fillna(data['Discount Applied'].mode()[0],inplace=True)
print("Discount Applied :",data['Discount Applied'].isnull().any())
```

**Data does not contain any data span so create one as "Warranty_Period"**

```python
import pandas as pd
import numpy as np
from datetime import timedelta


# Generate random Purchase_Date between Jan–Mar 2025
data['Purchase_Date'] = pd.to_datetime(
    pd.to_datetime('2025-01-01') + pd.to_timedelta(np.random.randint(0, 90, size=len(data)), unit='D')
)

# Delivery date = Purchase date + 3 to 10 days
data['Purchase_Date'] + pd.to_timedelta(np.random.randint(3, 11, size=len(data)), unit='D')

# Warranty period span as a string
data['Warranty_Period'] = data['Purchase_Date'].dt.strftime('%Y-%m-%d') + ' to ' + (
    data['Purchase_Date'] + pd.DateOffset(years=1)).dt.strftime('%Y-%m-%d')
```

```python
data
```

**Split the data into 2 section like start date and end date**

```python

data['First_date']=data['Warranty_Period'].str.split('to').str[0]
data
data['Last_date']=data['Warranty_Period'].str.split('to').str[1]
data
```
