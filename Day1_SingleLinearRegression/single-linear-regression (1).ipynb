```python
from sklearn.linear_model import LinearRegression

X=df[["Height(Inches)"]]
y=df["Weight(Pounds)"]

model=LinearRegression()
model.fit(X,y)
while True:

    user_input=input("enter height")
    if user_input.lower()=='exit':
        print("loop ended")
        break
    try:
        Height=float(user_input)
        s=model.predict([[Height]])
        print("predictied value is ",s)

    except ValueError:
        print("enter correct data")
        
```

✅ Formula of Linear Regression:

Predicted Y=mX+c
Where:

m = coefficient (also called slope)

c = intercept



```python
coef=model.coef_[0]
print("coeficient",coef)

inter=model.intercept_
print("intercept:",inter)
```
