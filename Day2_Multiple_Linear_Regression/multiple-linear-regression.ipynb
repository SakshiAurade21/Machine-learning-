```python
import pandas as pd
from sklearn.linear_model import LinearRegression


data=pd.read_csv("/kaggle/input/advertising-dataset/advertising.csv")
data.head(5)

```

```python
X=data[['TV','Radio','Newspaper']]
y=data.Sales
model=LinearRegression()
model.fit(X,y)
print("Model Trained sucessfully !")

print("coefficient :",model.coef_)
print("intercept :",model.intercept_)

y_pred=model.predict(X)

while True:
    TV=input("enter Tv data :")
    if TV.lower()=='exit':
        print("loop ended !")
        break
    Radio=input("enter Radio data :")
    Newspaper=input("enter Newspaper data :")

    try:
        TV=float(TV)
        Radio=float(Radio)
        Newspaper=float(Newspaper)
        user_input=[[TV,Radio,Newspaper]]
        predicated=model.predict(user_input)
        print("predicated Sales are :",predicated)
    except ValueError:
        print("galt logic dala hai !")

```

```python
import matplotlib.pyplot as plt
plt.scatter(y,y_pred,c='orange')
plt.plot([y.min(),y.max()],[y.min(),y.max()],c='green',linestyle=':')
plt.title("Actual vs Predicted")
plt.xlabel("Actual")
plt.ylabel("predicted")
```
